### Robustness check on benchmarks over Syrinx-based performance landscapes

This script conducts multiple simulations of the benchmark model, across learning rules and RL noise levels, and generates summary statistics.

```python
# Libraries

import numpy as np
import matplotlib.pyplot as plt               
from ipynb.fs.full.BenchmarkModel import Model as Model
import os
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
# Parameters

perf_threshold = 0.6                   # Success threshold
n_global_optima = 3                    # 1 for Gaussian-based landscapes and 3 for Syrinx-based landscapes

nrpd = 100                             # No. of simulations per scenario
n_repeat = nrpd + 1

nd = 60                                # No. of days in simulation
lr = 0.1                               # Learning rate of simulation
ntpd = 1000                            # No. of trials per day of simulation

noiseVal = [0.2, 0.3, 0.4, 0.5]        # RL noise levels to be tested

benchmark_type = 'DevRL'               # Benchmark earning rule to simulate (DevRL/StdRL/Annealing)

Fig_path = 'BenchmarkwSyrinx_' + str(nd) + 'd_0' + str(int(lr*100)) +'eta/' + benchmark_type + '/' 
```

```python
overall_rseed = 1000
np.random.seed(overall_rseed)

rseeds = np.random.randint(0, 1e9, size=n_repeat)

# Stores performance metrics of each simulation
Performance = {}
Performance['rseeds'] = rseeds 
```

```python
# Loads performance metrics of previously generated simulations

Performance = np.load(Fig_path + 'Performance.npy').item()
```

```python
# Plots summary figure of performance metrics across noise levels, for a given algorithm.

for noiseLim in noiseVal:
    fig_path = Fig_path + 'Noise_0'+ str(int(noiseLim*100)) + '/'
    
    PerfMetric = Performance[noiseLim*100]['RewardPerf']

    figure, ax = plt.subplots(1, 1)

    plt.title('Syrinx with ' + str(int(noiseLim*100)) + '% noise DevRL', y=1.15, x=0.5)

    for i in range(1):
        ax.plot(PerfMetric[i*nrpd:(i+1)*nrpd], lw=0, marker='.', color='purple')
        p = np.sum(PerfMetric[i*nrpd:(i+1)*nrpd]>perf_threshold) * 100 / nrpd
        ax.text(nrpd/2-.5, 1.1, str(p)+'%', color='purple', ha='center')

        ax.axhline(y=perf_threshold, color='grey', linestyle='--')

        ax.set_ylim(0,1)
    #     ax.set_xlabel(str((i+1)*10))

        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.spines['bottom'].set_visible(False)
        ax.spines['left'].set_visible(False)
        ax.get_xaxis().set_ticks([])
        ax.get_yaxis().set_ticks([])

    ax.set_ylabel('Performance')
    ax.spines['left'].set_visible(True)
    ax.get_yaxis().set_ticks([0, 1])
    # ax.set_title('Distractors', y=-0.15, x=0.5)


#     plt.savefig(fig_path + 'Overall_rewardPerf_' + str(int(noiseLim*100)) + '.png')
```

```python
# Plots summary statistics of benchmark model on Syrinx performance landscapes across noise levels, to build comparison across learning rules

benchmark_types = ['Dual', 'StdRL', 'DevRL', 'Annealing']
benchmark_performance_stats = [syrinx_model_Performance, syrinx_stdRL_Performance, syrinx_devRL_Performance, syrinx_annealing_Performance]

np.random.seed(0)
figure, ax = plt.subplots(1,len(benchmark_types))
axt = [ax[0].twinx(), ax[1].twinx(), ax[2].twinx(), ax[3].twinx()]


# Cmaps = ['plasma_r', 'plasma_r', 'plasma_r', 'plasma_r']
cmap='plasma'
Alpha = [.1, .2, .3, .4]

for i, benchmark in enumerate(benchmark_types):
    benchmark_Performance = benchmark_performance_stats[i]
    
    for k, noiseLim in enumerate(noiseVal):
        PerfMetric_benchmark = benchmark_Performance[noiseLim*100]['RewardPerf']
        Perf_x = np.ones(nrpd)*(k+1)*10 + np.random.uniform(-1,1,nrpd)
        Perf_y = PerfMetric_benchmark[:nrpd]
        h = np.histogram(Perf_y, bins=np.linspace(0,1,11))[0]
        z = h[(Perf_y*10).astype(int)]
        Perf_noiseLim = PerfMetric_benchmark[:nrpd]
        P_abovethres = Perf_noiseLim[Perf_noiseLim>perf_threshold]
        p = P_abovethres.size * 100 / nrpd
        
        axt[i].bar((k+1)*10, p, 5, bottom=0, align='center', edgecolor='black', color='grey', alpha=0.15*k)
        axt[i].bar((k+1)*10, p, 5, bottom=0, align='center', edgecolor='black', color='', alpha=.8)
        axt[i].text((k+1)*10-2, 5, str(int(p))+'%', color='black', ha='right', va='bottom', rotation=90, fontsize=18)
#         axt[i].bar((k+1)*10, p-100, 3, bottom=0, align='center', edgecolor='purple', color='', alpha=.2)


        ax[i].axhline(y=perf_threshold, color='grey', linestyle='--', label=('Success threshold'*(k==0)))
        
        im = ax[i].scatter(Perf_x, Perf_y, c='purple', s=30, alpha=.5, edgecolor='None', vmin=0, vmax=1, label=('Terminal performance'*(k==0)))
        ax[i].scatter((k+1)*10, np.median(Perf_y), s=180, facecolors='black', edgecolors='black', marker='X', alpha=1) 
        ax[i].scatter((k+1)*10, np.median(Perf_y), s=100, facecolors='yellow', edgecolors='black', marker='X', alpha=1, label=('Median'*(k==0))) 


    ax[i].set_ylim(0,1.2)
    ax[i].set_xlim(0,50)

    ax[i].spines['top'].set_visible(False)
    ax[i].spines['right'].set_visible(False)
    ax[i].spines['bottom'].set_visible(False)
    ax[i].spines['left'].set_visible(False)
    ax[i].get_xaxis().set_ticks([])
    ax[i].get_yaxis().set_ticks([])
    ax[i].get_xaxis().set_ticks([10, 20, 30, 40])
    ax[i].get_xaxis().set_ticklabels(['20%', '30%', '40%', '50%'], fontsize=20, color='black', rotation=45)
    
    ax[i].get_xticklabels()[0].set_alpha(0.7) 
    ax[i].get_xticklabels()[1].set_alpha(0.8) 
    ax[i].get_xticklabels()[2].set_alpha(0.9) 
    ax[i].get_xticklabels()[3].set_alpha(1) 
    ax[i].tick_params(axis='x', which='both', length=0)

    
    axt[i].spines['top'].set_visible(False)
    axt[i].spines['left'].set_visible(False)
    axt[i].spines['bottom'].set_visible(False)
    axt[i].spines['right'].set_visible(False)
    
    axt[i].set_ylim(-100,100)
    axt[i].get_yaxis().set_ticks([])

ax[0].spines['left'].set_visible(True)
ax[0].set_ylabel('Terminal performance', fontsize=25,y=.4)
ax[0].get_yaxis().set_ticks([0, .6, 1.00])
ax[0].spines['left'].set_bounds(0, 1)
ax[0].yaxis.label.set_color('Purple')
ax[0].spines['left'].set_color('Purple')
ax[0].tick_params(axis='y', colors='Purple', labelsize=15)
ax[0].get_yticklabels()[1].set_color('black') 
# ax[0].tick_params(labelsize=15)

    
ax[1].set_xlabel('Initial RL noise level', x=1.1, fontsize=20)
# ax[-1].legend(loc='lower right', frameon=False, fontsize=18)

axt[-1].spines['right'].set_visible(True)
axt[-1].spines['right'].set_bounds(0, 100)
axt[-1].get_yaxis().set_ticks([0, 100])
axt[-1].set_ylabel('Success rate', y=.75, rotation=270, fontsize=25)
axt[-1].tick_params(labelsize=15)
                            
ax[0].set_title('Dual\npathway\n(Dual)', y=1.1, va='center', ha='center', fontsize=22)
ax[1].set_title('Standard\nRL\n(StdRL)', y=1.1, va='center', ha='center', fontsize=22)
ax[2].set_title('RL with\ndecreasing noise\n(DevRL)', y=1.1, va='center', ha='center', fontsize=22)
ax[3].set_title('Simulated\nAnnealing\n(SA)', y=1.1, va='center', ha='center', fontsize=22)

# figure.suptitle('Syrinx contour')
# ax[0].tick_params(axis='y', colors='Purple')


# Uncomment all the lines below to save full scale image while using appropriate %matplotlib backend
figManager = plt.get_current_fig_manager()
figManager.window.showMaximized()

plt.savefig('Figures/' + 'Fig4b-Benchmarks' + '.png', bbox_inches="tight", rasterized=False)
```

### Statistical tests

Mann-whitney U test was performed to compare distributions of terminal performance between different learning rules.

```python
from scipy.stats import mannwhitneyu
```

```python
# Loads performances at 20\% noise level

dual_perf = syrinx_model_Performance[20]['RewardPerf']
stdrl_perf = syrinx_stdRL_Performance[20]['RewardPerf']
devrl_perf = syrinx_devRL_Performance[20]['RewardPerf']
annealing_perf = syrinx_annealing_Performance[20]['RewardPerf']
```

```python
np.median(dual_perf[:-1]),np.median(stdrl_perf[:-1])
```

```python
mannwhitneyu(dual_perf[:-1], stdrl_perf[:-1], alternative='two-sided')
```

```python
np.median(dual_perf[:-1]),np.median(annealing_perf[:-1])
```

```python
mannwhitneyu(dual_perf[:-1], annealing_perf[:-1], alternative='two-sided')
```

```python
np.median(dual_perf[:-1]),np.median(devrl_perf[:-1])
```

```python
mannwhitneyu(dual_perf[:-1], devrl_perf[:-1], alternative='two-sided')
```

```python
# Loads performances at 50\% noise level

dual_perf = syrinx_model_Performance[50]['RewardPerf']
devrl_perf = syrinx_devRL_Performance[50]['RewardPerf']
```

```python
np.median(dual_perf[:-1]),np.median(devrl_perf[:-1])
```

### Robustness check on benchmarks over Gaussian-based performance landscapes

This script conducts multiple simulations of the benchmark model, across learning rules, classes of landscapes and RL noise levels, and generates summary statistics.

```python
# Libraries

import numpy as np
import matplotlib.pyplot as plt                            
from ipynb.fs.full.BenchmarkModel import Model as Model
import os
```

```python
# Parameters

perf_threshold = 0.6                   # Success threshold
n_global_optima = 1                    # 1 for Gaussian-based landscapes and 3 for Syrinx-based landscapes

nrpd = 100                             # No. of simulations per scenario
n_distractors = 3                      # No. of landscapes classes (Low, Medium, High)
n_repeat = nrpd*n_distractors + 1

nd = 60                                # No. of days in simulation
lr = 0.1                               # Learning rate of simulation
ntpd = 1000                            # No. of trials per day of simulation

noiseVal = [0.2, 0.3, 0.4, 0.5]        # RL noise levels to be tested
distractor_vals = [5,40,160]           # No. of Gaussian hills to superimpose

benchmark_type = 'DevRL'               # Benchmark earning rule to simulate (DevRL/StdRL/Annealing)

Fig_path = 'BenchmarkwArtificial_' + str(nd) + 'd_0' + str(int(lr*100)) +'eta/' + benchmark_type + '/' 
```

```python
overall_rseed = 1000
np.random.seed(overall_rseed)

rseeds = np.random.randint(0, 1e9, size=n_repeat)

# Stores performance metrics of each simulation
Performance = {}
Performance['rseeds'] = rseeds 
```

```python
# Loads performance metrics of previously generated simulations

Performance = np.load(Fig_path + 'Performance.npy').item()
```

```python
# Plots summary figure of performance metrics across noise levels and landscape classes, for a given algorithm.

for noiseLim in noiseVal:
    fig_path = Fig_path + 'Noise_0'+ str(int(noiseLim*100)) + '/'
    
    PerfMetric = Performance[noiseLim*100]['RewardPerf']

    figure, ax = plt.subplots(1, n_distractors)
    if n_distractors == 1: ax = [ax]

    plt.title('Benchmark with ' + str(int(noiseLim*100)) + '% noise', y=1.15, x=-1)

    for i in range(n_distractors):
        ax[i].plot(PerfMetric[i*nrpd:(i+1)*nrpd], lw=0, marker='.', color='purple')
        p = np.sum(PerfMetric[i*nrpd:(i+1)*nrpd]>perf_threshold) * 100 / nrpd
        ax[i].text(nrpd/2-.5, 1.1, str(p)+'%', color='purple', ha='center')

        ax[i].axhline(y=perf_threshold, color='grey', linestyle='--')

        ax[i].set_ylim(0,1)
    #     ax.set_xlabel(str((i+1)*10))

        ax[i].spines['top'].set_visible(False)
        ax[i].spines['right'].set_visible(False)
        ax[i].spines['bottom'].set_visible(False)
        ax[i].spines['left'].set_visible(False)
        ax[i].get_xaxis().set_ticks([])
        ax[i].get_yaxis().set_ticks([])

    ax[0].set_ylabel('Performance')
    ax[0].spines['left'].set_visible(True)
    ax[0].get_yaxis().set_ticks([0, 1])
    # ax.set_title('Distractors', y=-0.15, x=0.5)


#     plt.savefig(fig_path + 'Overall_rewardPerf_' + str(int(noiseLim*100)) + '.png')
```

```python
# Plots summary figure of performance metrics across noise levels with a medium class landscape, for a given algorithm.

figure, ax = plt.subplots(1, len(noiseVal))
k = 0
for noiseLim in noiseVal:
    PerfMetric_dual = artificial_model_Performance[noiseLim*100]['RewardPerf']
    PerfMetric_stdRL = artificial_stdRL_Performance[noiseLim*100]['RewardPerf']
    PerfMetric_devRL = artificial_devRL_Performance[noiseLim*100]['RewardPerf']
    PerfMetric_annealing = artificial_annealing_Performance[noiseLim*100]['RewardPerf']

    i=1    
    
    
    j=0
#     for i in range(n_distractors):
    ax[k].plot(40*j+np.ones(nrpd)*(i+1)*10, PerfMetric_dual[i*nrpd:(i+1)*nrpd]*100, lw=0, marker='.', color='purple', alpha=0.4)
    p = np.sum(PerfMetric_dual[i*nrpd:(i+1)*nrpd]>perf_threshold) * 100 / nrpd
    ax[k].bar(40*j+(i+1)*10, p, width=20, color='grey', alpha=0.2)
    ax[k].text(40*j+(i+1)*10, 5, str(int(p))+'%', color='black', ha='center', rotation = 45)
    ax[k].text(40*j+20, -5, 'Dual', color='black', ha='left', rotation = 90)
      
    j=1
#     for i in range(n_distractors):
    ax[k].plot(40*j+np.ones(nrpd)*(i+1)*10, PerfMetric_stdRL[i*nrpd:(i+1)*nrpd]*100, lw=0, marker='.', color='purple', alpha=0.4)
    p = np.sum(PerfMetric_stdRL[i*nrpd:(i+1)*nrpd]>perf_threshold) * 100 / nrpd
    ax[k].bar(40*j+(i+1)*10, p, width=20, color='grey', alpha=0.2)
    ax[k].text(40*j+(i+1)*10, 5, str(int(p))+'%', color='black', ha='center', rotation = 45)
    ax[k].text(40*j+20, -5, 'StdRL', color='black', ha='left', rotation = 90)
        
    j=2
#     for i in range(n_distractors):
    ax[k].plot(40*j+np.ones(nrpd)*(i+1)*10, PerfMetric_annealing[i*nrpd:(i+1)*nrpd]*100, lw=0, marker='.', color='purple', alpha=0.4)
    p = np.sum(PerfMetric_annealing[i*nrpd:(i+1)*nrpd]>perf_threshold) * 100 / nrpd
    ax[k].bar(40*j+(i+1)*10, p, width=20, color='grey', alpha=0.2)
    ax[k].text(40*j+(i+1)*10, 5, str(int(p))+'%', color='black', ha='center', rotation = 45)
    ax[k].text(40*j+20, -5, 'Annealing', color='black', ha='left', rotation = 90)
        
    j=3
#     for i in range(n_distractors):
    ax[k].plot(40*j+np.ones(nrpd)*(i+1)*10, PerfMetric_devRL[i*nrpd:(i+1)*nrpd]*100, lw=0, marker='.', color='purple', alpha=0.4)
    p = np.sum(PerfMetric_devRL[i*nrpd:(i+1)*nrpd]>perf_threshold) * 100 / nrpd
    ax[k].bar(40*j+(i+1)*10, p, width=20, color='grey', alpha=0.2)
    ax[k].text(40*j+(i+1)*10, 5, str(int(p))+'%', color='black', ha='center', rotation = 45)
    ax[k].text(40*j+20, -5, 'DevRL', color='black', ha='left', rotation = 90)
        
    ax[k].set_ylim(0,100)

    ax[k].axhline(y=perf_threshold*100, color='grey', linestyle='--')

    ax[k].set_ylim(0,100)
    ax[k].set_title(str(int(noiseLim*100))+"% noise")

    ax[k].spines['top'].set_visible(False)
    ax[k].spines['right'].set_visible(False)
    ax[k].spines['bottom'].set_visible(False)
    ax[k].spines['left'].set_visible(False)
    ax[k].get_yaxis().set_ticks([])
    ax[k].get_xaxis().set_ticks([])
#     ax[k].get_xaxis().set_ticks([20, 60, 100, 140])
#     ax[k].get_xaxis().set_ticklabels([5, 25, 50, 5, 25, 50, 5, 25, 50, 5, 25, 50], rotation=90)        
    
    k = k + 1
        
        
ax[0].set_ylabel('Performance metric / success rate')
ax[0].get_yaxis().set_ticks([0, 25, 50, 75, 100])
ax[0].spines['left'].set_visible(True)
# plt.xlabel('Distractors', x=-0.1)
figure.suptitle('Artificial contour')
    
# plt.show()
# plt.savefig(Fig_path+'/benchmarkVsdual_25.png')
```
